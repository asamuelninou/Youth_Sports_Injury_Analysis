```python
import pandas as pd
neiss2010 = pd.read_csv("neiss2010_Alicia.csv")
neiss2011 = pd.read_csv("neiss2011_Alicia.csv")
neiss2012 = pd.read_csv("neiss2012_Alicia.csv")
neiss2013 = pd.read_csv("neiss2013_Alicia.csv")
neiss2014 = pd.read_csv("neiss2014_Alicia.csv")
neiss2015 = pd.read_csv("neiss2015_Alicia.csv")
neiss2016 = pd.read_csv("neiss2016_Alicia.csv")
neiss2017 = pd.read_csv("neiss2017_Alicia.csv")
neiss2018 = pd.read_csv("neiss2018_Alicia.csv")
neiss2019 = pd.read_csv("neiss2019_Alicia.csv")
```

```python
csv_files = [neiss2010,neiss2011,neiss2012,neiss2013,neiss2014,neiss2015,neiss2016, neiss2017, neiss2018, neiss2019]
```

```python
masterneiss = pd.concat(csv_files)
masterneiss.head()
```

```python
masterneiss_targetage = masterneiss[(masterneiss["Age"] > 14) & (masterneiss["Age"]< 25)]
```

```python
masterneiss_targetage_product = masterneiss_targetage[(masterneiss_targetage["Product_1"] == 1205) | 
                                            (masterneiss_targetage["Product_1"] == 1267) | 
                                            (masterneiss_targetage["Product_1"] == 1268) | 
                                            (masterneiss_targetage["Product_1"] == 3225) | 
                                            (masterneiss_targetage["Product_1"] == 3226) | 
                                            (masterneiss_targetage["Product_1"] == 3237) | 
                                            (masterneiss_targetage["Product_1"] == 3241) | 
                                            (masterneiss_targetage["Product_1"] == 3242) | 
                                            (masterneiss_targetage["Product_1"] == 3267) |
                                            (masterneiss_targetage["Product_1"] == 3271) |
                                            (masterneiss_targetage["Product_1"] == 5030) ]
```

```python
df = masterneiss_targetage_product.groupby(["Product_1"])["Product_1"].count()
```
