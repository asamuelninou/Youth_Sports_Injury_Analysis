```python
import pandas as pd 
import os 

neiss2019_df = pd.read_csv("../Data/neiss2019.csv")
neiss2019_df.head()
```

```python
#The first step to take in cleaning the data is going to be narrowing it down based on age. 
#As a team we decided that the age range should be 14-24 years old, so I will use this variable to do an initial cleanse. 

neiss2019_df = neiss2019_df.loc[(neiss2019_df["Age"] <= 24) & (neiss2019_df["Age"] >= 14)]

neiss2019_df.head(15)

```

```python
#The next step I'm going to take is cutting down the data based on specific injuries. 
#The injuries we chose to look at for this project are: sprains, concussions, broken bones, and dislocations

neiss2019_df = neiss2019_df.loc[ (neiss2019_df["Diagnosis"] == 64 ) | (neiss2019_df["Diagnosis"] == 55) |  (neiss2019_df["Diagnosis"] == 52) | (neiss2019_df["Diagnosis"] == 57) ]

neiss2019_df.head(10)


```

```python
#Check that it worked. 
df = neiss2019_df.groupby(["Diagnosis"])["Diagnosis"].count()

print(df)

```

```python
#That worked, now I'm going to drop columns we don't need. Specifically I'm going to drop other_race, treatment_date, fire_involvement, and all other variables we won't need. 

neiss2019_df = neiss2019_df.drop("Other_Race", axis= 1)
neiss2019_df = neiss2019_df.drop("Treatment_Date", axis= 1)
neiss2019_df = neiss2019_df.drop("Fire_Involvement", axis= 1)
neiss2019_df = neiss2019_df.drop("Hispanic", axis= 1)
neiss2019_df = neiss2019_df.drop("CPSC_Case_Number", axis= 1)
neiss2019_df = neiss2019_df.drop("Narrative", axis= 1)
neiss2019_df = neiss2019_df.drop("Other_Diagnosis", axis= 1)
neiss2019_df = neiss2019_df.drop("Body_Part_2", axis= 1)
neiss2019_df = neiss2019_df.drop("Diagnosis_2", axis= 1)
neiss2019_df = neiss2019_df.drop("Other_Diagnosis_2", axis= 1)
neiss2019_df = neiss2019_df.drop("Product_3", axis= 1)

neiss2019_df.head(10)
```

```python
#initial cleaning is done, so now this will be written into a csv file and sent to the cleaning group
#for further cleaning and visualization 

#make the csv file 

output_name = "../Data/neiss2019_clean.csv"

neiss2019_df.to_csv(output_name, index_label = 'Age')
```
